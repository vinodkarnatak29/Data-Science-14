```python
# Load datasets from CSV

# use Pandas read_csv() to load data
users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv' )
sessions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv' )
products =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv' )
transactions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
```

```python
#12. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

#Left Join to display all transactions with left join with Users
transactions.merge(users, how='left', on='UserID')
```

```python
#13. Which transactions have a UserID not in users

# To get data for those users which are not available in Transactions , get userID infromation from users table
#Data for those users , whose information is not available in Transaction table
transactions[~transactions.UserID.isin(users.UserID)]
```

```python
#14. Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)

#Use Join and display only matched records.
transactions.merge(users, how='inner', on='UserID')
```

```python
#15. Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

#Use outer Join and display both  matched and non matching records from user and tansactions table
transactions.merge(users, how='outer', on='UserID')
```

```python
#16. Determine which sessions occurred on the same day each user registered

#Join User ID and Session Date with Registed date and locate the matching rows.
pd.merge( users,sessions, right_on=['UserID','SessionDate'],left_on=['UserID','Registered'],how='inner')
```

```python
#17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

#create a user dataframe with key and user ID field
user_df= pd.DataFrame({'key': np.repeat(1, users.shape[0]), 'UserID': users.UserID})

#credte a Product  dataframe with key and Product  ID field
product_df = pd.DataFrame({'key': np.repeat(1, products.shape[0]), 'ProductID': products.ProductID})

#Merge both dataframae using KEY Value
df_users_products=pd.merge(user_df, product_df,on='key')[['UserID', 'ProductID']]
df_users_products
```

```python
#18. Determine how much quantity of each product was purchased by each user

# to get data for quantity for products which was purchased by users merge and group the data in users and transaction data based upon userID and ProductID & get sum of values in quantity column based upon pair of userId and productId (using apply() function) and fill na values with zero (0)
quantity_data_df=pd.merge(df_users_products, transactions, how='left', on=['UserID', 'ProductID']).groupby(['UserID', 'ProductID']).apply(lambda x: pd.Series(dict(Quantity=x.Quantity.sum()))).reset_index().fillna(0)

quantity_data_df
```

```python
#19. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)

#Merge Transaction usng USer ID.
pd.merge(transactions, transactions, on='UserID')
```

```python
#20. Join each user to his/her first occuring transaction in the transactions table

#Use Group by and Frist functions to locate first occuring transaction from Transactions and users dataframe.
pd.merge(users, transactions.groupby('UserID').first().reset_index(), how='left', on='UserID')
```

```python
#21. Test to see if we can drop columns

#create a new dataframe by merging users and transactions
data=users.merge(transactions, how='left',on='UserID')

#print the rows
data
```

```python
# Get cloumn's list convert dataframe columns into list
my_columns= list(data) 
my_columns
```

```python
#print the field which have more thatn 90# not null values
list(data.dropna(thresh=int(data.shape[0] *.9), axis=1).columns)
```

```python
#Print the columns which are having atleast one missed values and store in a variable
missing_info=list(data.columns[data.isnull().any()])
missing_info
```

```python
#Identify and Print Columns names with missed count
print('Output: Count of missing data')
print(' ')

#Loop through missing values column by one my one
for col in missing_info:
    #calcuate and store missed data count
    num_missing = (data[data[col].isnull() == True].shape[0]) 
    #Print the number of values missed in particular columns
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
#Identify and Print Columns names with missed count's percentage

print('Output of percentage missing data ')
print(' ')

#Loop through missing values column by one my one
for col in missing_info:
     #calcuate and store missed data percentate
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0] 
    #Print the %  missed in particular columns
    print('number missing for column {}: {}'.format(col, percent_missing))
```
